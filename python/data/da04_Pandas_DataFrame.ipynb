### DataFrame
#### DataFrame - 생성
    DataFrame은 2차원 배열형식, 표같은 스트레드시트 자료구조
    여러 개의 컬럼을 가지며 서로 다른 종류의 값이 담긴다.
    
    DataFrame은 다양한 방식으로 생성되지만 가장 흔하게 생성하는 방법
    1) 리스트를 값으로 가지는 딕셔너리 방법으로 생성
    2) Numpy 배열을 이용해서 생성
    3) read_csv(), read_excel()함수 등을 이용해서 생성

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
# 1. 딕셔너리를 이용해서 생성
df_dic = {
        "state":['Ohio','Ohio','Ohio','Nevada','Nevada','Nevada'],
        "year":[2000, 2001, 2002, 2001, 2002, 2003],
        'pop':[1.5, 1.7, 3.6, 2.4, 2.9, 3.2]
}

#시리즈의 결합체 -> 데이터프레임(서로 다른 타입들의 컬럼값을 가짐)
dicDf = DataFrame(df_dic)
dicDf
```

```python
print(type(dicDf.state))
print(type(dicDf.year))
print(type(dicDf.pop))
print(type(dicDf["pop"]))
```

```python
data1 = {'name': ['James','Peter','Jane','Tomas'],
       'address':["NY","NY","LA","Texas"],
       "age":[33,44,55,66]
      }

df1 = DataFrame(data1)
df1
```

```python
# 2. random 함수 사용하기
df2 = DataFrame(np.random.randint(10,100,16).reshape(4,4),
               columns = list('abcd'),
               index = list("abcd")) # 키워드 매개변수 -> 매개변수명을 알고 사용
                                     #                    순서는 상관 없음
    
# columns를 사용해서 컬럼명을 수정
df2.columns = ["one","two","three","four"]
df2
```

####               원본이 바로 변경                        원본변경이 안됨 inplace = True

```python
df2
```

```python
df2.columns = ["A-class","B-class","C-class","D-class"]
df2
```

```python
df1
```

```python
df1_1 = df1.rename(columns={"address":"addr"}) # shift + tab
#inplace = False => 원본 변경 안된다.
# 따라서 inplace = True => 원본이 변경됨
df1_1
```

### DataFrame - 조회하기

```python
df2.columns = ["one","two","three","four"]
df2
```

```python
# 슬라이싱 2가지 방법(숫자 슬라이싱, 라벨 슬라이싱)
# 숫자 슬라이싱
df2[0:2]
```

```python
# 라벨 슬라이싱
df2["a":"b"]
```

```python
# one, two 컬럼에 대한 모든 데이터를 조회
# 행부분에 one, two 없기 떄문에 헤딩만 출력
# 컬럼을 조회할 때는 슬라이싱 문법이 아니다.
# 컬럼을 조회할 때는 2차원으로 가져와야 함...
df2[["one","two"]] # 2차원으로 검색해야 한다.
```

```python
df2[df2["two"]>45][["two"]]
```

### DataFrame 조회하기 
iloc, loc<br>
iat, at -> 스칼라 값 조회(1 row, 1 column)

```python
df2[df2['two']>45].iloc[:,1:2]
```

```python
df1
```

```python
# 1. James, Peter 라인 전부 가져오기...iloc, loc
print(df1.iloc[0:2,0:4])
print("*"*30)
print(df1.iloc[:2,:4])
print("*"*30)
print(df1.iloc[:2])
print("*"*30)
print(df1.loc[0:1,"name":"age"]) # 라벨로 가져옴
```

```python
# 2. 스칼라값 가져오기 iat, at
# Peter의 주소, James의 나이
print(df1.iat[1,1])
print(df1.iat[0,2])

print("*"*30)
print(df1.at[1,"addr"])
print(df1.at[0,"age"])
```

```python
# 값 변경하기
df1.iat[2,2] = 50
df1
```

```python
# loc를 이용해서 마지막 줄에 결측값 넣기
df1.loc[4] = np.nan
df1
```

### Tips 실제 데이터로 응용하기

```python
# 1. 상위 3개 데이터 조회
df3.head(3)
```

```python
# 2. tip비율로 정렬(높은 팁이 먼저 나오도록...), 샘플 데이터 5개만 출력
df3.sort_values(by = ["tip"],ascending=False).head()
```

```python
df3.sort_values(by = ["tip"],ascending=False).head().plot(kind = "bar")
plt.show()
```

```python
df3.sort_values(by = ["tip"],ascending=False).head()["tip"].plot(kind = "bar")
plt.show()
```

```python
df3.sort_values(by = ["tip"],ascending=False).head()["tip"].plot(kind = "barh")
plt.show()
```

```python
# 3. 요일별, 팁별 데이터를 정렬하라. 이때, 요일별은 내림차순, 팁은 오름차순으로 정렬되로록
# 마지막으로 데이터는 끝에서 7개만 출력되도록

df3.sort_values(by=["day",'tip'], ascending=[False,True]).tail(7)
```
